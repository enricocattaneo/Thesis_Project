## **Create Dataset with Stats' Features** for both *Cups and Leagues* matches:
1) *Import **Packages***
2) *Import **Datasets***

5) **Merge** Original Data with *Stats Data* from point (1)

### 1) Import Packages

```python
# Import needed packages
import pandas as pd
import numpy as np
import collections
from tqdm.notebook import trange
# Change Pandas rows and columns' options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

#### Functions

```python
# Turn a nested dictionary into a flattened dictionary
def flatten(d, parent_key='', sep='_'): 
    items = []
    for k, v in d.items():
        new_key = parent_key + sep + k if parent_key else k
        if isinstance(v, collections.MutableMapping):
            items.extend(flatten(v, new_key, sep=sep).items())
        else:
            items.append((new_key, v))
    return dict(items)
```

### 2) Import Datasets

```python
# Import leagues_data from csv
leagues_data = pd.read_csv('Input Data/leagues_static.csv', low_memory = False)
leagues_data.set_index('id', inplace = True)
leagues_data.tail()
```

```python
# Import cups_data from csv
cups_data = pd.read_csv('Input Data/cups_static.csv', low_memory = False)
cups_data.set_index('id', inplace = True)
cups_data.tail()
```

### 3) Concatenate Vertically Leagues Dataset and Cups Dataset 

```python
# Control before concatenating 
print(len(leagues_data))
print(len(cups_data))

dataframes = [cups_data, leagues_data]
if all([set(dataframes[0].columns) == set(df.columns) for df in dataframes]):
    print('All have the same columns')
else:
    print('Some have different columns')
    
print(len(leagues_data.columns) == len(cups_data.columns))
```

```python
complete_data = pd.concat([leagues_data, cups_data], ignore_index=False)
print(len(complete_data) == len(leagues_data) + len(cups_data))
complete_data.tail()
```

### 4) Create Features from Stats Column

```python
all_stats = []
for i in trange(len(complete_data)):
    diz_single_stats = {}
    single_stats = eval(complete_data['stats'].iloc[i])
    if not single_stats or len(single_stats) == 1:
        all_stats.append(diz_single_stats)
    else:
        diz_single_stats['Home'] = single_stats[0]
        diz_single_stats['Away'] = single_stats[1]
        all_stats.append(flatten(diz_single_stats))

stats_df = pd.DataFrame(all_stats) 
stats_df.dropna(how = 'all', inplace = True)
print(len(stats_df) == len(complete_data)) 
stats_df.tail()
```

### 5) Merge Original Data with Stats Data

```python
all_df = pd.merge(complete_data.reset_index(), stats_df, how = 'outer', 
                         left_on = 'id', right_on = 'Away_fixture_id').set_index('id')
```

```python
print(len(complete_data) == len(all_df))
print(leagues_data.index.difference(all_df.index))
```

### 6) Drop Useless Columns

```python
features_to_drop = ['visitorCoach_firstname', 'season_current_round_id', 'league_coverage_predictions', 
                    'localCoach_common_name','neutral_venue', 'time_starting_at_timestamp', 
                    'league_coverage_topscorer_goals', 'visitorTeam_national_team','league_logo_path', 
                    'localTeam_is_placeholder', 'visitorCoach_image_path', 'aggregate_id', 'referee_firstname',
                    'details', 'time_starting_at_date', 'weather_report_icon', 'colors', 
                    'season_current_stage_id', 'localCoach_firstname', 'league_coverage_topscorer_cards', 
                    'winning_odds_calculated', 'scores_visitorteam_pen_score', 'league_is_friendly', 'league_active',
                    'localCoach_image_path', 'leg', 'referee_lastname', 'scores_ps_score', 'time_second', 
                    'time_extra_minute', 'visitorTeam_is_placeholder', 'weather_report_temperature_temp',
                    'group_id', 'league_type', 'visitorCoach_common_name', 'weather_report_temperature_unit', 
                    'localCoach_lastname', 'localTeam_logo_path', 'league_live_standings', 'weather_report', 
                    'league_coverage_topscorer_assists', 'time_starting_at_timezone', 'venue_image_path', 
                    'is_placeholder', 'pitch', 'scores_localteam_pen_score', 'scores_et_score', 'deleted', 
                    'referee_common_name', 'visitorCoach_lastname', 'visitorTeam_logo_path', 
                    'localTeam_national_team', 'commentaries', 'weather_report_temperature_celcius_unit', 
                    'colors_visitorteam_kit_colors', 'colors_localteam_kit_colors', 'venue_address', 
                    'league_legacy_id', 'visitorTeam_current_season_id', 'localTeam_current_season_id', 
                    'localTeam_legacy_id', 'visitorTeam_legacy_id', 'time_added_time', 'league_current_season_id', 
                    'league_current_round_id', 'league_current_stage_id', 'league_legacy_id', 'stats', 'time_starting_at_time',
                    'weather_report_updated_at']
all_df.drop(features_to_drop, axis = 1, inplace = True)

# Consider only games that were actually played 
all_df = all_df[all_df['time_status'] == 'FT']
all_df.drop('time_status', axis = 1, inplace = True)
```

### 7) Transform columns' values 

```python
# Change columns values 
all_df['weather_lat_lon'] = list(zip(all_df.weather_report_coordinates_lat, all_df.weather_report_coordinates_lon))
all_df.drop(['weather_report_coordinates_lat', 'weather_report_coordinates_lon'], axis = 1, inplace = True)
all_df['weather_report_humidity'] = all_df['weather_report_humidity'].str.replace('%', '')
all_df['weather_report_clouds'] = all_df['weather_report_clouds'].str.replace('%', '')
all_df['weather_report_wind_speed'] = all_df['weather_report_wind_speed'].str.replace(' m/s', '')
all_df['venue_coordinates'] = '(' + all_df['venue_coordinates'] + ')'
all_df['venue_surface'] = np.where(all_df['venue_surface'] == 'artificial turf', 0 ,1) # na = grass
all_df['season_is_current_season'] = np.where(all_df['season_is_current_season'] == True, 1 ,0)
all_df['league_is_cup'] = np.where(all_df['league_is_cup'] == False, 0, 1)
```

### 8) Rename Columns

```python
all_df.rename(columns={'weather_report_humidity': 'weather_humidity(%)','weather_report_clouds': 'weather_clouds(%)',
                        'weather_report_wind_speed': 'weather_windspeed(m/s)','venue_surface': 'venue_grass_surface',
                        'season_is_current_season': 'current_season'}, inplace=True)
all_df.columns = all_df.columns.str.lower().str.replace("localteam", "Home")
all_df.columns = all_df.columns.str.lower().str.replace("visitorteam", "Away")
all_df.columns = all_df.columns.str.lower().str.replace("local", "Home")
all_df.columns = all_df.columns.str.lower().str.replace("visitor", "Away")
all_df = all_df.loc[:,~all_df.columns.duplicated()]
```

### 9) Change columns data types

```python
float_columns = ['weather_report_pressure', 'weather_report_temperature_celcius_temp', 'weather_report_wind_degree', 'weather_windspeed(m/s)', 'home_passes_percentage', 'away_passes_percentage'] # columns that will be floats
# Convert all numeric columns to int except those that are expressed in floats
all_df[['weather_windspeed(m/s)', 'weather_clouds(%)', 'weather_humidity(%)']] = all_df[['weather_windspeed(m/s)', 'weather_clouds(%)', 'weather_humidity(%)']].astype('float64')       
m = all_df.select_dtypes(np.number).loc[:, ~all_df.select_dtypes(np.number).columns.isin(float_columns)]
all_df[m.columns]= m.round().astype('Int64')
all_df['time_starting_at_date_time'] = pd.to_datetime(all_df['time_starting_at_date_time'], infer_datetime_format=True)
all_df['round_start'] = pd.to_datetime(all_df['round_start'], format = '%Y-%m-%d')
all_df['round_end'] = pd.to_datetime(all_df['round_end'], format = '%Y-%m-%d')
```
